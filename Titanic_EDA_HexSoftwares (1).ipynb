## Data Cleaning

Data cleaning was performed to handle missing values and prepare the dataset for analysis. Numerical features were treated differently from categorical features to preserve data integrity.


```python
df['age'] = df['age'].fillna(df['age'].median())
```

## Exploratory Data Analysis (EDA)

In this section, visualizations are used to explore patterns, relationships, and trends within the Titanic dataset to understand factors influencing passenger survival.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

The survival distribution shows that a larger proportion of passengers did not survive the disaster. This highlights the severity of the Titanic incident and provides context for further analysis of survival factors.

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

Male passengers had a significantly higher survival rate compared to female passengers. This suggests that gender played an important role in survival, likely influenced by evacuation priorities such as women and children first.

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

The age distribution shows that most passengers were young to middle-aged adults. Younger passengers appear to have had a slightly higher chance of survival, indicating age may have influenced survival outcomes.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The correlation analysis reveals that passenger class and fare have noticeable relationships with survival. Higher-class passengers and those who paid higher fares were more likely to survive, suggesting socioeconomic status was a key factor.

Key insights

* Male passengers had a much higher survival rate than females

* Passengers in higher classes were more likely to survive

* Fare showed a positive relationship with survival

* Younger passengers had slightly better survival outcomes
